```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import rasterio
import cv2
from PIL import Image
```

```python
band_paths = []

for i in range(1, 12):
    path = f"/mnt/data/Optical/Landsat_8/LC08_L1TP_086241_20250830_20250830_02_RT/LC08_L1TP_086241_20250830_20250830_02_RT_B{i}.TIF"
    band_paths.append(path)

print(band_paths)
```

```python
band_paths
```

```python
band = []
for i in range(1, 12):
    with rasterio.open(band_paths[i-1]) as src:
        b = src.read(1)
        band.append(b)

plt.figure(figsize=(12, 20))
plt.imshow(band[0], cmap='gray')
plt.title(f"Band 1")
plt.axis('off')
plt.show()
```
