```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import rasterio
import cv2
from PIL import Image
```

```python
band_paths = []

for i in range(1, 12):
    path = f"/mnt/data/Optical/Landsat_8/LC08_L1TP_086241_20250830_20250830_02_RT/LC08_L1TP_086241_20250830_20250830_02_RT_B{i}.TIF"
    band_paths.append(path)
```

```python
band = []
for i in range(1, 12):
    with rasterio.open(band_paths[i-1]) as src:
        b = src.read(1)
        band.append(b)

plt.figure(figsize=(12, 12))
plt.imshow(band[0], cmap='gray')
plt.title(f"Band 1")
plt.axis('off')
plt.show()
```

```python
plt.figure(figsize=[10,5])
plt.hist(band[0].flatten(), bins=256*2,range= [5000,15000])
plt.show()
```

```python
plt.figure(figsize=[12,12])
plt.imshow(band[0], vmin=6000, vmax=11000,cmap = 'gray')
plt.colorbar()
```

```python
band_1_cropped = band[0][5000:6000,4000:4500].copy()
print(band_1_cropped.shape)
```

```python
plt.figure(figsize=[10,5])
plt.hist(band_1_cropped.flatten(), bins=256*2,range= [5000,15000])
plt.show()
```

```python
plt.figure(figsize=[12,12])
plt.imshow(band_1_cropped, vmin=6500, vmax=11000,cmap = 'gray')
plt.colorbar()
```
